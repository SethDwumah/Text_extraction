```python
import textract
from langchain.text_splitter import CharacterTextSplitter
import json
```

## Extract text from .pdf, .csv, .doc, .txt,.png,.jpg, .xlsx

```python
def convert(filename):
    text = textract.process(filename, encoding='utf-8')
    text = text.decode('utf-8').strip().replace("\n","")
    if not text:
        text = textract.process(filename, encoding='utf-8', method='tesseract')

    data ={'extracted_text':text}
    json_data = json.dumps(data,indent=4)

    with open("pdf_text.json","w") as json_file:
        json_file.write(json_data)

    return text
    
```

```python
text = convert("attention all you need.pdf") # the extention of the file is required
```

## Create chunks of text

```python
def get_text_chunks(text):
    text_splitter = CharacterTextSplitter(
        separator="\n",
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len
    )
    chunks = text_splitter.split_text(text.lower())
    return chunks
        
```

```python
get_text_chunks(text)
```
